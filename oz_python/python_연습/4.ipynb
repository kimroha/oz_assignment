내장함수 : divmod

> 내장함수란 ? : 자주 사용하는 기능을 미리 만들어 놓은것

**divmod** : 몫과 나머지를 동시에 구해줌

```python
print(divmod(3,2))
print(divmod(7,4))
print(divmod(13,2))
print(divmod(19,4))
print(divmod(15,5))
```
